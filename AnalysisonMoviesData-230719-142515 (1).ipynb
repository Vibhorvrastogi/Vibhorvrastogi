**QUES1) How have audience preferences for film genres evolved over the past decade?**

We'll focus on the Year and Genre columns to analyze how audience preferences for film genres have evolved over the past decade.


```python
Movie['Genre'] = Movie['Genre'].str.split(',')
exploded_movies_data = Movie.explode('Genre')
genre_trends = exploded_movies_data.groupby(['Year', 'Genre']).size().reset_index(name='Count')
genre_trends_pivot = genre_trends.pivot(index='Year', columns='Genre', values='Count').fillna(0)
genre_trends_pivot.head()
```

```python
import matplotlib.pyplot as plt
top_genres = genre_trends_pivot.sum().sort_values(ascending=False).head(10).index
plt.figure(figsize=(14, 8))
for genre in top_genres:
    plt.plot(genre_trends_pivot.index, genre_trends_pivot[genre], label=genre)

plt.title('Trends in Popular Film Genres Over the Past Decade')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.legend(title='Genre')
plt.grid(True)
plt.show()
```

Here are some observations:

1.)Action and Adventure: These genres have consistently been popular, with a noticeable increase in the number of movies released in recent years.
2.)Drama: This genre has also remained consistently popular, with many releases each year.
3.)Comedy: The number of comedy films has been relatively stable, with some fluctuations.
4.)Animation and Family: These genres show a steady presence, reflecting consistent demand.
5.)Horror and Thriller: These genres have shown some variability but have maintained a steady presence over the decade.

**Ques2.) What demographic factors correlate most strongly with box office success?**

The strongest correlation with box office success is the number of votes a movie receives, followed by its runtime and IMDb rating. These factors might reflect audience interest and engagement, contributing to higher revenue.

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style='whitegrid')


fig, axes = plt.subplots(1, 3, figsize=(18, 5))


sns.scatterplot(data=Movie, x='Votes', y='Revenue (Millions)', ax=axes[0])
sns.regplot(data=Movie, x='Votes', y='Revenue (Millions)', scatter=False, ax=axes[0], color='red')
axes[0].set_title('Revenue vs. Votes')


sns.scatterplot(data=Movie, x='Runtime (Minutes)', y='Revenue (Millions)', ax=axes[1])
sns.regplot(data=Movie, x='Runtime (Minutes)', y='Revenue (Millions)', scatter=False, ax=axes[1], color='red')
axes[1].set_title('Revenue vs. Runtime')


sns.scatterplot(data=Movie, x='Rating', y='Revenue (Millions)', ax=axes[2])
sns.regplot(data=Movie, x='Rating', y='Revenue (Millions)', scatter=False, ax=axes[2], color='red')
axes[2].set_title('Revenue vs. Rating')


for ax in axes:
    ax.set_xlabel('')
    ax.set_ylabel('Revenue (Millions)')


plt.tight_layout()


plt.show()
```

**Ques.3) How do critical reviews and audience ratings influence box office performance?**



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Movie, x='Metascore', y='Revenue (Millions)')
plt.title('Revenue vs. Metascore')
plt.xlabel('Metascore')
plt.ylabel('Revenue (Millions)')
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(data=Movie, x='Rating', y='Revenue (Millions)')
plt.title('Revenue vs. Rating')
plt.xlabel('Rating')
plt.ylabel('Revenue (Millions)')
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(data=Movie, x='Votes', y='Revenue (Millions)')
plt.title('Revenue vs. Votes')
plt.xlabel('Votes')
plt.ylabel('Revenue (Millions)')
plt.grid(True)
plt.show()


correlations = Movie[['Revenue (Millions)', 'Metascore', 'Rating', 'Votes']].corr()


plt.figure(figsize=(10, 6))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

**Ques.4) Does the release year of a film has any impact on its overall box office performance**

To analyze the impact of the release year on box office performance, we can use the "Year" and "Revenue (Millions)" columns



```python
average_revenue_per_year = Movie.groupby('Year')['Revenue (Millions)'].mean()

# Plotting the average revenue per year
plt.figure(figsize=(10, 6))
average_revenue_per_year.plot(kind='line', marker='o', color='orange')
plt.title('Average Box Office Revenue by Release Year')
plt.xlabel('Year')
plt.ylabel('Average Revenue (Millions)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Ques5.) Does the combination of director and Actors affect the box office performance?**

To analyze the impact of the combination of directors and actors on box office performance, we can examine patterns in revenue associated with specific director-actor collaborations.

```python
Movie['Main Actor'] = Movie['Actors'].str.split(',').str[0]

# Creating a new column that combines director and main actor
Movie['Director-Actor Combo'] = Movie['Director'] + ' & ' + Movie['Main Actor']

# Grouping by the director-actor combination and calculating the average revenue
combo_revenue = Movie.groupby('Director-Actor Combo')['Revenue (Millions)'].mean().sort_values(ascending=False)

# Displaying the top 10 director-actor combinations by average revenue
combo_revenue.head(10)
# Plotting the top director-actor combinations by average revenue
plt.figure(figsize=(12, 8))
combo_revenue.head(10).plot(kind='barh', color='skyblue')
plt.title('Top Director-Actor Combinations by Average Box Office Revenue')
plt.xlabel('Average Revenue (Millions)')
plt.ylabel('Director-Actor Combination')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()  # Invert y-axis to have the highest value at the top
plt.tight_layout()
plt.show()
```

### **9.1 Summary and Conclusion**

Influence of Critical Reviews and Audience Ratings on Box Office Performance:

1.)Correlation Analysis: The analysis showed that there is a positive correlation between IMDb ratings and box office revenue, but it is relatively weak (0.218). The number of votes, representing audience engagement, has a stronger correlation with revenue (0.640). Metascore, which reflects critical reviews, has the weakest correlation with revenue (0.142).

2.)Scatter Plots: Visualizations confirmed these relationships, indicating that while critical reviews do have some influence, audience engagement (as indicated by votes) is a more significant predictor of box office success.

Optimal Runtime for Different Genres in Terms of Audience Satisfaction:

1.)Data Analysis: Movies were categorized by primary genre and runtime intervals to analyze average IMDb ratings. The findings showed that certain genres have higher average ratings within specific runtime ranges.

2.)Chart Visualization: The chart illustrated that audience satisfaction varies by genre and runtime. For example, action movies tend to have higher ratings in the 120-180-minute range, while shorter runtimes (<90 minutes) generally correlate with lower ratings across most genres.

Conclusion:

Audience Engagement Over Critical Acclaim: Box office performance is more closely linked to audience engagement than critical acclaim. Movies that generate more votes tend to perform better financially, suggesting that marketing and audience reach are crucial factors.

Optimal Runtime Varies by Genre: There is no one-size-fits-all optimal runtime for movies. However, different genres appear to have sweet spots in terms of runtime that correlate with higher audience satisfaction. For instance, action and adventure genres seem to perform well in terms of ratings with longer runtimes, whereas other genres may not follow the same pattern.

These insights can inform movie studios and producers about the potential financial and audience satisfaction impacts of critical reviews, audience engagement strategies, and film runtime decisions.
